```python

# write a function that can return the track record of 2 teams against each other

def track_record(t1, t2):
    c1 = (ipl['Team1'] == t1) & (ipl['Team2'] == t2)
    c2 = (ipl['Team1'] == t2) & (ipl['Team2'] == t1)

    matches_between_teams = ipl[c1 | c2]

    t1_wins = matches_between_teams['WinningTeam'] == t1
    t2_wins = matches_between_teams['WinningTeam'] == t2

    count1 = t1_wins.value_counts().get(True, 0)
    count2 = t2_wins.value_counts().get(True, 0)

    print(f"{t1} win {count1} matches")
    print(f"{t2} win {count2} matches")

t1 = input("Enter first team: ")
t2 = input("Enter second team: ")

track_record(t1, t2)
```
